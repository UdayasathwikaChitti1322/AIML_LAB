```python
import pandas as pd
import matplotlib.pyplot as plt
people=['kiran','arun','vijay','varun']
age=[11,12,13,14]
height=[3,4,5,6]
weight=[45,55,60,65]
data={
    'Name':people,
    'Age':age,
    'Height':height,
    'Weight':weight

}
df=pd.DataFrame(data)
print(df)
#scatterplot
plt.figure(figsize=(6,4))
plt.scatter(df['Age'],df['Weight'],color='red',marker='o')
plt.title('Scatter plot of Age vs Weight')
plt.xlabel('Age(years)')
plt.ylabel('Weight(kgs)')
#barplot
plt.figure(figsize=(8,6))
plt.title("Bar chart of individuals\Heights")
plt.bar(df['Name'],df['Height'],color='green')
plt.xlabel('Name')
plt.ylabel('Height(cm)')
#histogram
plt.figure(figsize=(7,5))
plt.hist(df['Age'],color='red',edgecolor='black',bins=8)
plt.title('Histogram of age distribution')
plt.xlabel('Age(years)')
plt.show()
```
